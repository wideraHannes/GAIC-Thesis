```python
import json
import pandas as pd

# Load results
with open('c1_reasoning_30_gpt-4_1-2025-04-14.json') as f:
    reasoning = json.load(f)
with open('c1_no_reasoning_30_gpt-4_1-2025-04-14.json') as f:
    no_reasoning = json.load(f)

# Extract F1 scores per dataset
rows = []
for ds in no_reasoning['datasets']:
    r = reasoning['datasets'].get(ds, {})
    nr = no_reasoning['datasets'][ds]
    rows.append({
        'Dataset': ds,
        'F1 (reasoning)': r.get('macro_f1_original'),
        'F1 (no reasoning)': nr.get('macro_f1_original'),
        'Delta': (r.get('macro_f1_original', 0) or 0) - (nr.get('macro_f1_original', 0) or 0)
    })

df = pd.DataFrame(rows)
df.loc['Mean'] = df.select_dtypes('number').mean()
df.loc['Mean', 'Dataset'] = 'MEAN'
df.style.format({'F1 (reasoning)': '{:.3f}', 'F1 (no reasoning)': '{:.3f}', 'Delta': '{:+.3f}'}).background_gradient(subset=['Delta'], cmap='RdYlGn')
```
